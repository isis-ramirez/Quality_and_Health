```python
#finding null values
df.isnull().sum()
```

```python
#replacing missing values with mean value
mean_value=df['GDP'].mean()
df['GDP'].fillna(value=mean_value, inplace=True)
print(df)
```

```python
#checking the GDP missing value count
df.isnull().sum()
```

```python
#replacing missing values with mean value for population
mean_value=df['Population'].mean()
df['Population'].fillna(value=mean_value, inplace=True)
print(df)
```

```python
#checking the population missing value count
df.isnull().sum()
```

```python
#deleting the columns we do not need
df=df.drop(df.columns[[6,7,13,18,19]], axis=1)
```

```python
#checking if columns got deleted
print(df)
```

```python
#final look at missing values count
df.isnull().sum()
```
